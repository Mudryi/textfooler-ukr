```python
import re
import pandas as pd

def parse_synonym_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        content = f.read()

    # Split entries by 'text {id}'
    entries = re.split(r'\btext (\d+)', content)[1:]  # Skip empty start
    
    parsed_data = []
    
    # Process in chunks of [id, entry]
    for i in range(0, len(entries), 2):
        text_id = int(entries[i])
        entry = entries[i+1]
        
        # Extract original sentence
        orig_match = re.search(r'orig sent \(\d+\):\s*(.+?)\s*adv sent', entry, re.DOTALL)
        original_sentence = orig_match.group(1).strip() if orig_match else None
        
        # Extract adversarial sentence
        adv_match = re.search(r'adv sent \(\d+\):\s*(.+?)\s*Replacements', entry, re.DOTALL)
        adversarial_sentence = adv_match.group(1).strip() if adv_match else None
        
        # Extract replacements list
        replacements_match = re.search(r"Replacements\s+\[(.*?)\]", entry, re.DOTALL)
        replacements_raw = replacements_match.group(1).replace("\n", " ") if replacements_match else ""
        replacements = re.findall(r"\('(.+?)',\s*'(.+?)',\s*\d+\)", replacements_raw)
        
        parsed_data.append({
            "text_id": text_id,
            "original_sentence": original_sentence,
            "adversarial_sentence": adversarial_sentence,
            "replacements": replacements
        })
    
    return pd.DataFrame(parsed_data)

df = parse_synonym_file("adv_results_reviews_sentence_transformer/adversaries.txt")
df.head()
```
